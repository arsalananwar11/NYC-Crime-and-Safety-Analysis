```python
import requests
import json
import pandas as pd
import functools as ft
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_endpoints_dict = {
    'Shooting Data' : 'https://data.cityofnewyork.us/resource/833y-fsy8.json',
    'Complaint Map' : 'https://data.cityofnewyork.us/resource/2fra-mtpn.json',
    'Use of Force' : 'https://data.cityofnewyork.us/resource/f4tj-796d.json'
}
```

```python
class Loader:
    def load_file(self, api_endpoint):
        data = requests.get(api_endpoint)
        data = json.loads(data.text)
        data_df = pd.json_normalize(data)
        data = None
        return data_df
        
```

```python
data_loader = Loader()
```

```python
shooting_data = data_loader.load_file(data_endpoints_dict['Shooting Data'])
shooting_data.head(5)
```

```python
shooting_data.boro = shooting_data.boro.str.title()
shooting_data.boro.unique()
```

```python
shooting_data = shooting_data.rename(columns={'boro': 'borough', 'occur_date': 'date'})
shooting_data.head(5)
```

```python
complaint_map_data = data_loader.load_file(data_endpoints_dict['Complaint Map'])
complaint_map_data.head(5)
```

```python
complaint_map_data.boro_nm = complaint_map_data.boro_nm.str.title()
complaint_map_data.boro_nm.unique()
```

```python
complaint_map_data = complaint_map_data.rename(columns={'boro_nm': 'borough', 'cmplnt_fr_dt': 'date'})
complaint_map_data.head(5)
```

```python
use_of_force_data = data_loader.load_file(data_endpoints_dict['Use of Force'])
use_of_force_data.head(5)
```

```python
use_of_force_data.patrol_borough.unique()
```

```python
borough_mapping = {
    'PBBX': 'Bronx', 
    'PBSI': 'Staten Island', 
    'PBMN': 'Manhattan', 
    'PBMS': 'Manhattan',
    'PBBN': 'Brooklyn', 
    'PBBS': 'Brooklyn', 
    'PBQS': 'Queens', 
    'PBQN': 'Queens'
}
```

```python
use_of_force_data.patrol_borough = use_of_force_data.patrol_borough.map(borough_mapping)
print(use_of_force_data.patrol_borough.unique())
use_of_force_data.head(5)
```

```python
use_of_force_data = use_of_force_data.rename(columns={'patrol_borough': 'borough', 'occurrence_date': 'date'})
use_of_force_data.head(5)
```

```python
dfs = [shooting_data, complaint_map_data, use_of_force_data]
df_final = ft.reduce(lambda left, right: pd.merge(left, right, on=['borough']), dfs)
df_final.head()
```
